```python
# Predicting with the trained YOLO model

# Import necessary modules
from ultralytics import YOLO
from PIL import Image
import os

# Define the file name of the image to be tested
FILENAME = 'yolo_dataset/images/val/random_shapes_48.png'

# Open and resize the image
img = Image.open(FILENAME)
img = img.resize([500, 500])

# Load the trained YOLO model (make sure to update the path if necessary)
model = YOLO("runs/detect/yolo/weights/best.pt")

# Perform prediction on the test image
res = model.predict(img, save=True)

# Open and display the predicted image
Image.open(os.path.join(res[0].save_dir,'image0.jpg'))
```
